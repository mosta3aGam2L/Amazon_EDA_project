```python
def detect_outliers_IQR(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

columns = ['discounted_price', 'actual_price', 'rating']

for col in columns:
    outliers, lb, ub = detect_outliers_IQR(df, col)
    print(f"{col} Outliers: {len(outliers)} rows")
    print(f"Lower Bound: {lb}, Upper Bound: {ub}\n")
```

## insights ##
- actual_price

Number of outliers: 213 rows

Minimum acceptable value: -4442.5 (illogical)

Maximum acceptable value: 9537.5

 Any product priced over 9537 is considered an outlier
 - discounted_price

Number of outliers: 217 rows

Minimum acceptable value: -2186 (illogical → any negative price is an error)

Maximum acceptable value: 4510

 Any product priced over 4510 is considered an outlier.
 - Any rating < 3.55 or > 4.75 is an outlier

```python
#Explain whether they should be removed or keptHandle incorrect or inconsistent data
df = df[(df['rating'] >= 0) & (df['rating'] <= 5)] 
df = df[(df['actual_price'] > 0) & (df['discounted_price'] > 0)]
```
