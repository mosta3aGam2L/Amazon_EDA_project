# 2-Data cleaning

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df = df.dropna(subset=['rating', 'rating_count'])
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

# convert data from string to (int ,float)
clean 3 columns and convert string to float , remove $  


```python
df['discounted_price'] = df['discounted_price'].str.replace('₹','').str.replace(',','').astype(float)
```

```python
df = df.dropna(subset=['actual_price'])
```

```python
df['actual_price'] = df['actual_price'].str.split('-').str[0]

df['actual_price'] = pd.to_numeric(
    df['actual_price'].str.replace('₹','').str.replace(',',''),
    errors='coerce'
)
```

syntax {  errors='coerce' }
->    
mean =any value not (int ,float) convert to nan value

```python
df['rating_count'] = df['rating_count'].str.replace(',','').astype(float)
```

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
```

# 3-Descriptive Statistics

1️⃣ Summary statistic

```python
df.describe()
```

```python
df['category'].value_counts()
```

//2️⃣category Distribution

```python
top = df['category'].value_counts().head(10).index

plt.figure(figsize=(10,6))
sns.countplot(data=df, x='category', order=top)
plt.xticks(rotation=45)
plt.title("Top 10 Product Categories")
plt.show()
```

syntax{plt.xticks(rotation=45) } 

use to draw and show name of category

3️⃣ Price statistics

```python
#Discount price
print("Discounted Price Stats:")
print(df['discounted_price'].describe())
#actual price
print("\nActual Price Stats:")
print(df['actual_price'].describe())
#discount percentage
df['discount_percentage'] = ((df['actual_price'] - df['discounted_price']) / df['actual_price'])*100
print("\nDiscount Percentage Stats:")
print(df['discount_percentage'].describe())
```

4️⃣ Rating statistics

```python
# coutn of rating
print("Rating Count Stats:")
print(df['rating_count'].describe())

# Average rating by category
rating_by_category = df.groupby('category')['rating_count'].mean().sort_values(ascending=False)
print("\nAverage Ratings by Category:")
print(rating_by_category)
```

```python
plt.figure(figsize=(10,5))
sns.histplot(df['discounted_price'], bins=20, kde=True)
plt.title("Discounted Price Distribution")
plt.show()

plt.figure(figsize=(10,5))
sns.histplot(df['rating_count'], bins=20, kde=True)
plt.title("Rating Count Distribution")
plt.show()
```

# 4-Data Visualization

```python
top_categories = df['category'].value_counts().head(10)
short_names = [name[:15] + '...' if len(name)>15 else name for name in top_categories.index]

plt.figure(figsize=(12,6))
sns.barplot(x=short_names, y=top_categories.values)
plt.xticks(rotation=45)
plt.title("Top 10 Product Categories (Short Names)")
plt.xlabel("Category")
plt.ylabel("Number of Products")
plt.show()
```
