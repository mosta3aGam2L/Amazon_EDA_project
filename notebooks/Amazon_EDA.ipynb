# 2-Data cleaning

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df = df.dropna(subset=['rating', 'rating_count'])
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

# convert data from string to (int ,float)
clean 3 columns and convert string to float , remove $  


```python
df['discounted_price'] = df['discounted_price'].str.replace('₹','').str.replace(',','').astype(float)
```

```python
df = df.dropna(subset=['actual_price'])
```

```python
df['actual_price'] = df['actual_price'].str.split('-').str[0]

df['actual_price'] = pd.to_numeric(
    df['actual_price'].str.replace('₹','').str.replace(',',''),
    errors='coerce'
)
```

syntax {  errors='coerce' }
->    
mean =any value not (int ,float) convert to nan value

```python
df['rating_count'] = df['rating_count'].str.replace(',','').astype(float)
```

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
```

```python
# Business Question: هل فيه منتجات بأسعار غير طبيعية (مرتفعة جداً أو منخفضة جداً)؟

fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Box plot للأسعار
axes[0, 0].boxplot(df['actual_price'].dropna())
axes[0, 0].set_title('Price Distribution - Box Plot', fontsize=12, fontweight='bold')
axes[0, 0].set_ylabel('Price ($)')
axes[0, 0].set_xticklabels(['Actual Price'])

# Histogram
axes[0, 1].hist(df['actual_price'].dropna(), bins=50, edgecolor='black', alpha=0.7)
axes[0, 1].set_title('Price Distribution - Histogram', fontsize=12, fontweight='bold')
axes[0, 1].set_xlabel('Price ($)')
axes[0, 1].set_ylabel('Frequency')
axes[0, 1].axvline(df['actual_price'].mean(), color='red', linestyle='--', label=f'Mean: ${df["actual_price"].mean():.2f}')
axes[0, 1].axvline(df['actual_price'].median(), color='green', linestyle='--', label=f'Median: ${df["actual_price"].median():.2f}')
axes[0, 1].legend()

# IQR Method
Q1 = df['actual_price'].quantile(0.25)
Q3 = df['actual_price'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

price_outliers = df[(df['actual_price'] < lower_bound) | (df['actual_price'] > upper_bound)]
price_outliers_high = df[df['actual_price'] > upper_bound].nlargest(10, 'actual_price')
price_outliers_low = df[df['actual_price'] < lower_bound].nsmallest(10, 'actual_price')

# Scatter plot مع الـ thresholds
axes[1, 0].scatter(range(len(df)), df['actual_price'], alpha=0.5, s=10)
axes[1, 0].axhline(y=upper_bound, color='red', linestyle='--', label=f'Upper Bound: ${upper_bound:.2f}')
axes[1, 0].axhline(y=lower_bound, color='red', linestyle='--', label=f'Lower Bound: ${lower_bound:.2f}')
axes[1, 0].set_title(f'Price Outliers Detection (IQR Method)\n{len(price_outliers)} outliers found ({len(price_outliers)/len(df)*100:.2f}%)', fontsize=12, fontweight='bold')
axes[1, 0].set_xlabel('Product Index')
axes[1, 0].set_ylabel('Price ($)')
axes[1, 0].legend()
axes[1, 0].set_yscale('log')  # Log scale عشان نشوف التوزيع أفضل

# Top outliers details
outlier_stats = pd.DataFrame({
    'Metric': ['Min Price', 'Max Price', 'Mean Price', 'Median Price', 'Q1', 'Q3', 'IQR', 'Lower Bound', 'Upper Bound', 'Outliers Count', 'Outliers %'],
    'Value': [
        f"${df['actual_price'].min():.2f}",
        f"${df['actual_price'].max():.2f}",
        f"${df['actual_price'].mean():.2f}",
        f"${df['actual_price'].median():.2f}",
        f"${Q1:.2f}",
        f"${Q3:.2f}",
        f"${IQR:.2f}",
        f"${lower_bound:.2f}",
        f"${upper_bound:.2f}",
        len(price_outliers),
        f"{len(price_outliers)/len(df)*100:.2f}%"
    ]
})

axes[1, 1].axis('tight')
axes[1, 1].axis('off')
table = axes[1, 1].table(cellText=outlier_stats.values, colLabels=outlier_stats.columns, cellLoc='left', loc='center')
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.2, 1.5)
axes[1, 1].set_title('Price Outliers Statistics', fontsize=12, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
# ----------- Detect Outliers باستخدام IQR -----------
Q1 = df['actual_price'].quantile(0.25)
Q3 = df['actual_price'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# ----------- Data بعد إزالة Outliers -----------
df_clean = df[(df['actual_price'] >= lower_bound) & (df['actual_price'] <= upper_bound)]

# ----------- Plot -----------
plt.figure(figsize=(14,6))

# قبل إزالة الـ outliers
plt.subplot(1,2,1)
sns.boxplot(x=df['actual_price'], color='orange')
plt.title("Before Removing Outliers", fontsize=14, fontweight='bold')
plt.xlabel("Actual Price")

# بعد إزالة الـ outliers
plt.subplot(1,2,2)
sns.boxplot(x=df_clean['actual_price'], color='green')
plt.title("After Removing Outliers", fontsize=14, fontweight='bold')
plt.xlabel("Actual Price")

plt.tight_layout()
plt.show()
```

```python

# ---------------- Prices Outliers Detection ----------------
Q1_price = df['actual_price'].quantile(0.25)
Q3_price = df['actual_price'].quantile(0.75)
IQR_price = Q3_price - Q1_price
lower_price = Q1_price - 1.5 * IQR_price
upper_price = Q3_price + 1.5 * IQR_price

# Filtered DataFrame (remove extreme price outliers)
df_price_clean = df[(df['actual_price'] >= lower_price) & (df['actual_price'] <= upper_price)]

# ---------------- Ratings Outliers Detection ----------------
# Ratings normally between 1 and 5
df_rating_clean = df_price_clean[(df_price_clean['rating'] >= 1) & (df_price_clean['rating'] <= 5)]

# ---------------- Visualization ----------------
plt.figure(figsize=(16,6))

# Price Before
plt.subplot(2,2,1)
sns.boxplot(x=df['actual_price'], color='orange')
plt.title("Prices Before Removing Outliers", fontsize=12, fontweight='bold')
plt.xlabel("Actual Price")

# Price After
plt.subplot(2,2,2)
sns.boxplot(x=df_price_clean['actual_price'], color='green')
plt.title("Prices After Removing Outliers", fontsize=12, fontweight='bold')
plt.xlabel("Actual Price")

# Rating Before
plt.subplot(2,2,3)
sns.boxplot(x=df['rating'], color='purple')
plt.title("Ratings Before Cleaning", fontsize=12, fontweight='bold')
plt.xlabel("Rating")

# Rating After
plt.subplot(2,2,4)
sns.boxplot(x=df_rating_clean['rating'], color='blue')
plt.title("Ratings After Cleaning", fontsize=12, fontweight='bold')
plt.xlabel("Rating")

plt.tight_layout()
plt.show()
```
